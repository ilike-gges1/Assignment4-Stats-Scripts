# 3.1.2. Hypothesis testing: comparing two groups
Using SciPy for basic statistical tests

```python
from scipy import stats
```

## 3.1.2.1 Student's t-test: the simplest statistical test

### 1-sample t-test: testing the value of a popluation mean

```python
stats.ttest_1samp(data['VIQ'], 0) #sees if mean value is equal to zero. Since the p value is 10^-28, it is most definitely not zero
```

```python
stats.ttest_1samp(data['VIQ'], 110) #sees if mean value is equal to 110. It is a higher p-value so it could be 110
```

### 2-sample t-test: testing for difference across populations

```python
female_viq = data[data['Gender'] == 'Female']['VIQ'] #Just female VIQ data
male_viq = data[data['Gender'] == 'Male']['VIQ'] #Just male VIQ data
stats.ttest_ind(female_viq, male_viq) #2-sample TTest to see. Higher p-value indicates no real difference.
```

## 3.1.2.2. Paired tests: repeated measures on the same individuals

```python
stats.ttest_ind(data['FSIQ'], data['PIQ']) #2-sample test
```

```python
stats.ttest_rel(data['FSIQ'], data['PIQ']) #paired test
```

```python
stats.ttest_1samp(data['FSIQ'] - data['PIQ'], 0) #equivalent to paired, it is a 1 sample teset on difference
```

```python
stats.wilcoxon(data['FSIQ'], data['PIQ']) #Wilcoxon signed-rank test that relaxes the assumption of Gaussian errors
```

# 3.1.3. Linear models, multiple factors, and analysis of variance
Creating linear regressions, comparing multiple types of variables

## 3.1.3.1. "formulas" to specify statistical models in Python

```python
x = np.linspace(-5, 5, 20) #creating the x in the data frame
np.random.seed(1) #I am not really sure what this does
y = -5 + 3*x + 4 * np.random.normal (size=x.shape) #creating the y for the data gframe
data = pandas.DataFrame({'x': x, 'y': y}) #creating the data frame
```

```python
from statsmodels.formula.api import ols
model = ols("y~x", data).fit() #creating the linear regression model
```

```python
print(model.summary()) #printing the model summary
```

```python
print(model.params) #prints the intercept and the slope
```

## Categorical variables: comparing groups or multiple categories

```python
data = pandas.read_csv('brain_size.csv', sep=';', na_values=".") #comparing IQ of males and females
model = ols("VIQ ~ Gender + 1", data).fit()
print(model.summary()) 
```

```python
data_fisq = pandas.DataFrame({'iq': data['FSIQ'], 'type': 'fsiq'})
data_piq = pandas.DataFrame({'iq': data['PIQ'], 'type': 'piq'})
data_long = pandas.concat((data_fisq, data_piq))
print(data_long) #creating list of IQs indicated by categorical variable
```

```python
model = ols("iq ~ type", data_long).fit() #creating a linear regression of this model
print(model.summary())
```

```python
stats.ttest_ind(data['FSIQ'], data['PIQ']) #high p value little difference
```

## 3.1.3.2. Multiple Regression: including multiple factors

```python
data = pandas.read_csv('iris.csv') #new CSV just dropped
model = ols('sepal_width ~ name + petal_length', data).fit() #systematic effect of sepal and petal size
print(model.summary())
```

## 3.1.3.3. Post-hoc hypothesis testing: analysis of variance (ANOVA)

```python
print(model.f_test([0, 1, -1, 0])) #comparing with f test, above 0.05 not significant
```

```python
import pandas
from statsmodels.formula.api import ols

data = pandas.read_csv('../brain_size.csv', sep=';', na_values='.')

model = ols('VIQ ~ Gender + MRI_Count + Height', data).fit()
print(model.summary())
print(model.f_test([0, 1, 0, 0])) #high p-value, male and female not different after removing effect of brain size, height, weight
```

## 3.1.4. More visualization: seaborn for statistical exploration

```python
import urllib
import os

import matplotlib.pyplot as plt
import pandas

if not os.path.exists('wages.txt'):
    # Download the file if it is not present
    urllib.request.urlretrieve('http://lib.stat.cmu.edu/datasets/CPS_85_Wages',
                       'wages.txt')

# Give names to the columns
names = [
    'EDUCATION: Number of years of education',
    'SOUTH: 1=Person lives in South, 0=Person lives elsewhere',
    'SEX: 1=Female, 0=Male',
    'EXPERIENCE: Number of years of work experience',
    'UNION: 1=Union member, 0=Not union member',
    'WAGE: Wage (dollars per hour)',
    'AGE: years',
    'RACE: 1=Other, 2=Hispanic, 3=White',
    'OCCUPATION: 1=Management, 2=Sales, 3=Clerical, 4=Service, 5=Professional, 6=Other',
    'SECTOR: 0=Other, 1=Manufacturing, 2=Construction',
    'MARR: 0=Unmarried,  1=Married',
]

short_names = [n.split(':')[0] for n in names]

data = pandas.read_csv('wages.txt', skiprows=27, skipfooter=6, sep=None,
                       header=None, engine='python')
data.columns = short_names

import numpy as np
data['WAGE'] = np.log10(data['WAGE'])
```

```python
print(data)
```
